```python
from logger import logger
import numpy as np
from dumpers import SCV_Dumper
from multiprocessing import Queue, Process


class Test1:
    @logger.log(headers=["fun_name", "output", "args"])
    def foo(self, a, b):
        return a + b

    @logger.log()
    def boo(self):
        return (list([1, 23, 4]), np.array([1, 2, 3]))


class Test2:
    @logger.log()
    def freez(self, a=np.array([1, 234, 4]), b=4):
        return a + b

    @logger.log()
    def brah(self, c):
        return (list([1, 23, 4]), np.array([1, 2, 3]))
```

```python
dumper = SCV_Dumper(logger)
```

```python
test1 = Test1()
test2 = Test2()
```

```python
dupmer = Process(target=dumper.dump_loop)
dupmer.start()
```

```python
def experiments1():
    for i in range(4):
        test1.foo(i, i**2)
        test1.boo()
        test2.freez()


def experiments2():
    for _ in range(8):
        test2.brah(None)


Process(target=experiments1).start()
Process(target=experiments2).start()
```

```python
dupmer.terminate()
logger.get_logger_performance()
```
